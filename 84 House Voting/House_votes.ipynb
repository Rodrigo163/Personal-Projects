# House Votes analysis

Practice analysis from Congressional Voting Records Data Set 


https://archive.ics.uci.edu/ml/datasets/congressional+voting+records

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
from sklearn.neighbors import KNeighborsClassifier
```

```python
df = pd.read_csv('house-votes-84.csv.txt')
```

```python
#adding labels to columns
names = ["Party", "HI", "WP", "AB","PH", "ES", "RG", "AS", "ANI", "MM", "IM", "SYN", "EDU", "SUP", "CRI", "DUT", "EXPO"]
df.columns = names
```

```python
df.head()
```

```python
#if we wanted to clean the data we could kill all rows with the "?"
#In this case after doing this we end up with only half of the data so we'll keep the "?" as abstinence
#The inplace stuff makes sure that we're modifying the original df
#df.replace('?', np.NaN, inplace= True)
#df.dropna(inplace= True)
```

```python
df.replace(['n','y','?'], [0,1,2], inplace = True)
df.head()
```

```python
x = df.drop('Party', axis=1).values
y = df['Party'].values
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.25, random_state=42)
```

```python
#1st predition using KNN
knn = KNeighborsClassifier(n_neighbors = 6)
knn.fit(x_train, y_train)
knn.score(x_test, y_test)
```
