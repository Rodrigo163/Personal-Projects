# OpenCV Contours
https://opencv-python-tutroals.readthedocs.io/en/latest/py_tutorials/py_imgproc/py_contours/py_contours_begin/py_contours_begin.html

Contours can be explained simply as a curve joining all the continuous points (along the boundary), having same color or intensity. The contours are a useful tool for shape analysis and object detection and recognition.

```python
import numpy as np
import cv2
import matplotlib.pyplot as plt

def imshow(im):
    plt.imshow(im)
    plt.axis('off')
    plt.show()
```

```python
im = cv2.imread('test_leaf.jpeg')
imgray = cv2.cvtColor(im,cv2.COLOR_BGR2GRAY)
ret,thresh = cv2.threshold(imgray,127,255,0)
image, contours, hierarchy = cv2.findContours(thresh,cv2.RETR_TREE,cv2.CHAIN_APPROX_SIMPLE)
```

```python
im = cv2.drawContours(im, contours, 4, (0,255,0), 3)
imshow(im)
```
